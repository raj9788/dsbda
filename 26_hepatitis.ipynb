```python
def outliers(df,drop=False):
    for feature in df.columns:
        data = df[feature]
        Q1=np.percentile(data,25.)
        Q3=np.percentile(data,75.)
        IQR = Q3-Q1
        step = IQR*1.5
        outliers = data[~((data>=Q1-step)&(data<=Q3+step))].index.tolist()
        if not drop:
            print('For the feature {}, No of Outliers is {}'.format(feature, len(outliers)))
        if drop:
            df.drop(outliers, inplace = True, errors = 'ignore')
            print('Outliers from {} feature removed'.format(feature))
           
# def outliers(df_out, drop = False):
#   for each_feature in df_out.columns:
#       feature_data = df_out[each_feature]
#       Q1 = np.percentile(feature_data, 25.) # 25th percentile of the data of the given feature
#       Q3 = np.percentile(feature_data, 75.) # 75th percentile of the data of the given feature
#       IQR = Q3-Q1 #Interquartile Range
#       outlier_step = IQR * 2 #That's we were talking about above
#       outliers = feature_data[~((feature_data >= Q1 - outlier_step) & 
#       (feature_data <= Q3 + outlier_step))].index.tolist() 
#       if not drop:
#         print('For the feature {}, No of Outliers is {}'.format(each_feature, len(outliers)))
#       if drop:
#         df.drop(outliers, inplace = True, errors = 'ignore')
#         print('Outliers from {} feature removed'.format(each_feature))
```

```python
feature = ['Class'
,'AGE'
,'SEX'
,'STEROID'
,'ANTIVIRALS'
,'FATIGUE'
,'MALAISE'
,'ANOREXIA'
,'LIVER BIG'
,'LIVER FIRM'
,'SPLEEN PALPABLE'
,'SPIDERS'
,'ASCITES'
,'VARICES'
,'BILIRUBIN'
,'ALK PHOSPHATE'
,'SGOT'
,'ALBUMIN'
,'PROTIME'
,'HISTOLOGY']
```

```python
outliers(df[feature])
```

```python
x=df[feature]
y=df['Class']
df.shape
```

```python
x.drop(['Class'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=125)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn import metrics
print("accuracy  :: ",metrics.accuracy_score(y_pred,y_test))
print("precision :: ",metrics.precision_score(y_pred,y_test))
print("recall    :: ",metrics.recall_score(y_pred,y_test))
print("f1        :: ",metrics.f1_score(y_pred,y_test))
```
