```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.replace('?',np.nan,inplace=True)
```

```python
df.isna().sum()
```

```python
df.fillna(df.median(),inplace=True)
df.isna().sum()
```

```python
df.dtypes
for col in df.columns:
    df[col]=pd.to_numeric(df[col])
```

```python
x=df['num'].value_counts()
```

```python
plt.pie(x,labels=['very less chance','low chance','moderate chance','hifh chance','very high chance'],autopct="%1.1f%%")
```

```python
data=df['age']
plt.hist(data,color='green')
plt.xlabel("Age")
plt.ylabel("Frequency")
```

```python
data1=df['thalach']
data2=df['age']
data3=df['trestbps']
sns.boxplot(data=[data1,data2,data3])
```
